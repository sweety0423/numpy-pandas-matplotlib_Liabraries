# Arithmetic operations in pyhton pandas

```python
import pandas as pd
```

```python
var=pd.DataFrame({"a":[1,2,3,4,5,6],"b":[6,7,8,9,0,2]})
var
```

```python
#addition
var["c"]=var["a"]+var["b"]
var
```

```python
#substracton
var["c"]=var["a"]-var["b"]
var
```

```python
var["c"]=var["a"]/var["b"]
var
```

```python
var["c"]=var["a"]*var["b"]
var
```

```python
# logical operations
var1=pd.DataFrame({"a":[10,20,30,40,50,60],"b":[60,70,80,90,100,20]})
var1
```

```python
var1["python"]=var1['a']<=20 #creating a new column using logical operations
var1["python_1"]=var1['b']>=16
var1
```

# delete and insert functions in python pandas

## Insert

```python
var=pd.DataFrame({"a":[1,2,3,4,5],"b":[9,8,7,6,5]})
var
```

```python
var.insert(1,"PYHTON",var["a"]) # PARAMETERS(INDEX NO)(POSITON NO OR new column NAME)(DATA which you want to insert)
var # new column python with dublicate copy of a column
```

```python
var["python_1"]=var["a"][:3]
var
```

# delete the data

```python
#pop function is use to delete the column and shows the data which is deleted
var1=var.pop("python_1")
var1
```
